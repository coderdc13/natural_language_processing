## 6. Conclusion (NER) (6/6)
NLP offers many tools for text analysis. First, NLP provides a way to tabulate specific words in any given text; with this tabulation, the researcher may compose a specific and measured approach to the characterization of represented ideas within the given text. Furthermore, NLP gives the opportunity to break ideas into the smallest available units (which are often in the form of individual words) through the tokenization capability. In addition, NLP offers the opportunity to group words and count words according to specialized needs. Also, NLP offers the construction of graphics through the Word Cloud. The NER capability of NLP permits for the search, retrieval, and classification of concepts, words, and idioms. These capabilities lead the researcher to the need to study and master NLP use. In terms of discrepancies, the Application Programming Interfaces (APIs) change, so there can be no comparison.
